```python
# MATPLOTLIB OBJECT ORIENTED METHOD
```

```python
# CREATE FIGURE (EMPTY CANVAS)
fig=plt.figure()

axes=fig.add_axes([2.1,2.2,1.8,1.5])

axes.plot(x,y,'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

axes2.plot(y,x,'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
tips=sns.load_dataset("tips")
tips.head()
```

```python
plt.scatter(tips["total_bill"],tips["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.title("Bills vs Tips")
plt.show()
```

```python
# DISTPLOT
```

```python
sns.distplot(tips['total_bill']);
# safe to ignore warnings
#plt.boxplot(tips["total_bill"],showmeans=True);
```

```python
sns.distplot(tips['tip']);
# safe to ignore warnings
#plt.boxplot(tips["total_bill"],showmeans=True);
```

```python
sns.distplot(tips['total_bill'],kde=False)
#plt.boxplot(tips["tip"])
#sales in thousands , price in thousands and fuel efficiency
```

```python
# JOINTPLOT
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex');
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter');
```

```python
#PAIRPLOT
```

```python
sns.pairplot(tips);
```

```python
sns.pairplot(tips,hue='smoker');
```

```python
plt.figure(figsize=(15,15))
sns.pairplot(tips,hue='sex',palette="coolwarm");
#palette="husl"/"rainbow"/"coolwarm"
```

```python
# RUGPLOT
```

```python
sns.rugplot(tips['total_bill'])
sns.rugplot(tips["tip"])
```
