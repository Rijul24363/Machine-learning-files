```python
left1 = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                        'A':['A0','A1','A2','A3'],
                        'C':['B0','B1','B2','B3']})

```

```python
right1 = pd.DataFrame({'Key':['K0','K1','K2','K3'],
                        'C':['C0','C1','C2','C3'],
                        'D':['D0','D1','D2','D3']})
```

```python
left1
```

```python
right1
```

```python
# MERGING
```

```python
print(left1)
right1
```

```python
pd.merge(left1,right1,how="outer",on="Key")
#outer - It will print both the common observations as well as unique values from both the dataframes.
#left - it will only prints the data from left dataframe and common observations from the right table as well.
#right - it willonly prints the data from the right dataframe and common observations from the left table as well.
#inner - it will prints the common observations from both the tables.
```

```python
left = pd.DataFrame({'Key1':['K0','K0','K1','K2'],
                     'Key2':['K0','K1','K0','K1'],
                        '^':['^0','^1','^2','^3'],
                        'B':['B0','B1','B2','B3'],
                       "Age":[25,26,22,23],
                       "Sal":[250,260,220,230]})

right = pd.DataFrame({'Key1':['K0','K1','K1','K2'],
                              'Key2':['K0','K0','K0','K1'],
                                'C': ['C0','C1','C2','C3'],
                                'D': ['D0','D1','D2','D3'],
                                    "Age":[30,26,22,27],
                                    "Sal":[300,260,220,270]})
```

```python
print(left,"   left")
print()
print(right,"  right")
```

```python
#pd.concat([left right],axis=1)
#pd.concat([left,right])
pd.merge(left,right,on=["Key1","Key2"])
pd.merge(left,right,how="outer",on=["Key1","Key2"])
```

```python
print(left,"left Dataframe")
print()
print(right,"right DataFrame")
```
