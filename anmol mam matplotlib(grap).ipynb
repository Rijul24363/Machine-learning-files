```python
import matplotlib.pyplot as plt   #pyplot is a module inside matplotlib
plt.plot([1,2,3,4])      # bydefault it assumes that it is y axis
plt.ylabel('y axis')      # ylabel() to give label to yaxis
plt.xlabel('x axis')      ## xlabel() to give label to xaxis
plt.title("Line graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('dark_background')  # use dark_background instead of ggplot

x = [5,8,10]
y = [12,16,6]

x2 = [6,9,11]
y2 = [6,15,7]

# can plot specifically, after just showing the defaults:
plt.plot(x,y,'r',linewidth=5,label="red")
plt.plot(x2,y2,'b',linewidth=15,label="blue")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]
plt.stackplot(days, sleeping,eating,working,playing)

plt.xlabel('days')
plt.ylabel('activities')
plt.title('stackplot Graph')

plt.show()
```
