```python
import os
import numpy as np
import matplotlib.pyplot as plt

x = ['INFO', 'CUISINE', 'TYPE_OF_PLACE', 'DRINK', 'PLACE', 'MEAL_TIME', 'DISH', 'NEIGHBOURHOOD']
y = [160, 167, 137, 18, 120, 36, 155, 130]

fig, ax = plt.subplots(figsize=(20, 10))   
width = 0.75 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
print(ind)
plt.bar(ind, y, width=0.75, color="blue")
ax.set_xticks(ind)
ax.set_xticklabels(x, minor=False, size=20,rotation=90)
plt.title('title',fontsize=50)
plt.xlabel('x',fontsize=30)
plt.ylabel('y',fontsize=30)  

for i, v in enumerate(y):
    print(v)
    ax.text(i , v+2 , str(v), color='red', fontweight='bold')
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

x = [u'INFO', u'CUISINE', u'TYPE_OF_PLACE', u'DRINK', u'PLACE', u'MEAL_TIME', u'DISH', u'NEIGHBOURHOOD']
y = [160, 167, 137, 18, 120, 36, 155, 130]

fig, ax = plt.subplots(figsize=(20, 10))   
width = 0.75 # the width of the bars 
ind = np.arange(len(y))  # the x locations for the groups
plt.scatter(ind, y, width, color="blue")
ax.set_xticks(ind+width/10)
ax.set_xticklabels(x, minor=False, size=10,rotation=90)
plt.title('title')
plt.xlabel('x')
plt.ylabel('y')  

for i, v in enumerate(y):
    print(i,v)
    ax.text(i , v + 2, str((x[i],v)), color='blue', fontweight='bold')
plt.show()
```
