```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("C:/Users/rijul/Downloads/iris.csv")
```

```python
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
X = iris.iloc[:,:-1]
X
y = iris.iloc[:,-1]
#y
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state = 51)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
```

```python
RF = RandomForestClassifier(n_estimators=21,random_state=42)
```

```python
clf = RF.fit(X_train,y_train)
```

```python
clf.estimators_
```

```python
# VISUALIZING THE RANDOM FOREST
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf.estimators_[4],filled=True);
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf.estimators_[20],filled=True);
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf.estimators_[0],filled=True);
```

```python
# PREDICTION WITH RANDOM FOREST
```

```python
pred = RF.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test,pred)
accuracy_score(y_test,pred)
```
