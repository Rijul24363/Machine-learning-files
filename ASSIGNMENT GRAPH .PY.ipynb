```python
x=[10,15,80,25,30,35,40]
y=[12,35,78,16,18,32,73]
z=[4,8,9,11,34,67,90]
plt.hist(x)
plt.hist(y)
```

```python
star=plt.figure()
plt.plot(x, y,"yellow") # 'r','o'/'orange' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('First MatplotLib Graph')
plt.show()
```

```python
plt.scatter(y,z)
plt.xlabel("X Axis Title Here")
plt.ylabel("Y Axis Title Here")
plt.title("Second Matplot Graph")
plt.show()
```

```python
plt.subplot(151)
plt.tight_layout()
plt.plot(x, y, 'r--') # More on color options later

plt.subplot(1,5,2)
plt.tight_layout()
plt.plot(y, x, 'g^-',)

plt.subplot(1,5,3)
plt.tight_layout()
plt.plot(x,z,'y*-')

plt.subplot(1,5,4)
plt.tight_layout()
plt.plot(z,y,"c*-")#cyan

plt.subplot(1,5,5)
plt.tight_layout()
plt.plot(x, y, 'o-')

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([2.3,1.8,3.5,2.8])

axes.plot(x, y, 'b')
axes.set_xlabel('X Axis')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
plt.show()
```

```python

fig = plt.figure()

axes1 = fig.add_axes([1.1, 2.2, 3.3, 4.4])
axes2 = fig.add_axes([4.4, 3.3, 2.2, 1.1]) 

axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```
