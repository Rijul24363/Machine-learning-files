```python
import matplotlib.pyplot as plt   
# data to display on plots 
x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]  
# This will plot a simple scatter chart
plt.scatter(x, y)  
# Adding legend to the plot
plt.legend("y-axis data")

# Title to the plot
plt.title("Scatter chart")
  
plt.show()
```

```python
import matplotlib.pyplot as plt   
# data to display on plots 
x = [3, 1, 3, 12, 2, 4, 4] 
y = [3, 2, 1, 4, 5, 6, 7]   
# This will plot a simple bar chart
plt.bar(x, y)  
# Title to the plot
plt.title("Bar Chart")  
# Adding the legends
plt.legend(["bar"])
  
plt.show()
```

```python
import matplotlib.pyplot as plt   
# data to display on plots 
x = [1, 2, 3, 4]   
# this will explode the 1st wedge
# i.e. will seperate the 1st wedge
# from the chart
e  =(0.1, 0, 0, 0)  
# This will plot a simple pie chart
plt.pie(x, explode = e)  
# Title to the plot
plt.title("Pie chart")
  
plt.show()
```

```python
import matplotlib.pyplot as plt 
# data to display on plots 
x = [3, 1, 3] 
y = [3, 2, 1]   
# This will plot a simple line chart
# with elements of x as x axis and y
# as y axis
plt.plot(x, y) 
plt.title("Line Chart")  
# Adding the legends
plt.legend(["Line"])
  
plt.show()
```

```python
import matplotlib.pyplot as plt   
# data to display on plots 
x = [1, 2, 3, 4, 5, 6, 7, 4]   
# This will plot a simple histogram
plt.hist(x, bins = [1, 2, 3, 4, 5, 6, 7])  
# Title to the plot
plt.title("Histogram")  
# Adding the legends
plt.legend(["bar"])
  
plt.show()
```
