```python
from xgboost import XGBClassifier
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
from sklearn.datasets import load_breast_cancer
data=load_breast_cancer()
X=data["data"]
X
y=data["target"]
y
```

```python
data
```

```python
X=pd.DataFrame(X,columns=data.feature_names)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train
```

```python
model=XGBClassifier()
model.fit(X_train,y_train)
pred=model.predict(X_test)

training_pred=model.predict(X_train)

print("Accuracy of Training Data is ",accuracy_score(y_train,training_pred))

print("Accuracy of Test Data is ",accuracy_score(y_test,pred))
#confusion_matrix(y_test,pred)
```

```python
#XGBOOST Classifier with Hyper Parameter Tuning

model=XGBClassifier(max_depth=2,
                   subsample=0.4,#Range 0.1-1,subsample only lets to train 70% of observation.
                   n_estimators=51,#No. of Decision Trees
                   learning_rate=0.02,#Range 0.01-0.1
                   min_child_weight=1,
                   reg_alpha=0)#Penalizes the features that increases the error
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
y_train_pred=model.predict(X_train)
print("Training Data Accuracy ",accuracy_score(y_train,y_train_pred))
print()
print("Test Data Accuracy ",accuracy_score(y_test,y_pred))
```
